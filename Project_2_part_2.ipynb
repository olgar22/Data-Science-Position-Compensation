```python
url = 'https://en.wikipedia.org/wiki/List_of_capitals_in_the_United_States'
tables = pd.read_html(url)
df = tables[0]
cols = [1]
df = df[df.columns[cols]]
df.drop(df.index[:2], inplace=True)
df.reset_index(inplace = True)
df.head()
```

```python
years_exp = ['0','1','3','5','10','15','20']
list_of_records = []
for years_exp in years_exp:
    for state in df[1]:
        for job in  path['jobs']:
            #record = job+years_exp
            record_in = {
            'job': job,
            'state': state,
            'years': years_exp,
            'sal_min': '0',
            'sal_avg': '0'
            }
            record_out = data_scrape(**record_in)
            list_of_records.append(record_out)
            print(record_out)
        print('State ' + state + 'done')
print('Received ' + str(len(list_of_records)) + ' records')
            
```

```python
from csv import DictWriter
with open('spreadsheet.csv','w') as outfile:
    writer = DictWriter(outfile, ('job','state','years','sal_min','sal_avg'))
    writer.writeheader()
    writer.writerows(list_of_records)
```
